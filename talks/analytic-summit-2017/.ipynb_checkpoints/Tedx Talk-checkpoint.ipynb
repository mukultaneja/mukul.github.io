```python
import ast
import datetime
import matplotlib
import numpy as np
import pandas as pd
from wordcloud import WordCloud
from __future__ import division
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('data/ted_main.csv')

data.columns.tolist()
```

```python
data.info()
```

```python
data['talkyear'] = data['film_date'].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y'))
data['talkdate'] = data['film_date'].apply(lambda x: datetime.datetime.fromtimestamp(x).strftime('%Y-%m-%d'))
data['talkyear'] = data['talkyear'].apply(lambda x: int(x))
data['talkminduration'] = data['duration'] / 60
df = data.copy()

speakers = df['main_speaker'].str.split(',', expand=True).stack()
speakers = speakers.str.split('+', expand=True).stack()
speakers = speakers.str.split(' and ', expand=True).stack()
speakers = speakers.apply(lambda row: row.strip())
i = speakers.index.get_level_values(0)
df = df.loc[i].copy()
df['main_speaker'] = speakers.values
data = df

# data = data[data['talkyear'] >= 1999]

def wordCloud(col, val):
    words = list()
    tags = data[data[col] == val]['tags']
    tags = tags.apply(lambda x: ast.literal_eval(x))
    tags.apply(lambda x: words.extend(x))
    wordcloud = WordCloud(width = 500, height = 500).generate(' '.join(words))
    return wordcloud

def tagWordCloud(words):
    wordcloud = WordCloud(width = 500, height = 500).generate(' '.join(words))
    return wordcloud

def calc(row):
    temp, total = dict(), 0
    for _ in row:
        total += _['count']
    for _ in row:
        temp[_['name']] = '{0:.2f}'.format(_['count'] / total * 100)

    return temp
```

```python
df = data.groupby(['talkyear']).size().reset_index()
df.columns = ['talkyear', 'speakers']
df = df.set_index('talkyear')
df = df.sort_values(by=['speakers'], ascending=True)
df.plot.barh(figsize=(9, 9))
```

```python
df = data.groupby(['speaker_occupation']).size().reset_index()
df.columns = ['occupation', 'speakers']
df = df.sort_values(by=['speakers'], ascending=False).head(10)
df = df.set_index('occupation')
df = df.sort_values(by=['speakers'], ascending=True)
df.plot.barh()
```

```python
occupations = data.groupby(['speaker_occupation']).size().sort_values(ascending=False).reset_index().head(20)
occupations = occupations['speaker_occupation'].tolist()
lo, wpf, count = 20, 0, 5
while count <= lo:
    fig, ax = plt.subplots(1, 5, figsize=(20, 20))
    for i, v in enumerate(range(wpf,count)):
        wordcloud = wordCloud('speaker_occupation', occupations[v])
        ax[i].imshow(wordcloud)
        ax[i].axis('off')
        ax[i].text(0.5, -0.15, occupations[v] , size=15, ha="center", transform=ax[i].transAxes) 
    wpf = count
    count += 5
```

```python
talkyear = data.groupby(['talkyear']).size().reset_index().sort_values(by=['talkyear'], ascending=True)
talkyear = talkyear['talkyear'].tolist()

lo, wpf, count = 24, 0, 4
while count <= lo:
    fig, ax = plt.subplots(1, 4, figsize=(20, 20))
    for i, v in enumerate(range(wpf,count)):
        wordcloud = wordCloud('talkyear', talkyear[v])
        ax[i].imshow(wordcloud)
        ax[i].axis('off')
        ax[i].text(0.5, -0.15, talkyear[v] , size=15, ha="center", transform=ax[i].transAxes) 
    wpf = count
    count += 4
```

```python
speakers = raw_input().split(',')
l = len(speakers)
fig, ax = plt.subplots(1, l, figsize=(12, 5), sharey=True, sharex=True)
fig1, ax1 = plt.subplots(1, l, figsize=(15, 5))
colors = ["#1f77b4", "#aec7e8", "#ff7f0e", "#ffbb78", "#2ca02c", "#98df8a",
          "#d62728", "#ff9896", "#9467bd", "#c5b0d5", "#8c564b", "#c49c94",
          "#e377c2", "#f7b6d2", "#7f7f7f", "#c7c7c7", "#bcbd22", "#dbdb8d",
          "#17becf", "#9edae5"]

for index, speaker in enumerate(speakers):
    speaker = speaker.strip()
    words = list()
    ratings = dict()
    sd = data[data['main_speaker'] == speaker][['tags', 'ratings', 'title']].reset_index()

    # tags
    sd.apply(lambda row: words.extend(ast.literal_eval(row['tags'])), axis=1)

    # ratings
    sd['ratings'] = sd.apply(lambda row: calc(ast.literal_eval(row['ratings'])), axis=1)
    rl = len(sd['ratings'])
    keys = sd['ratings'].iloc[0].keys()
    for key in keys:
        s = 0
        for i in xrange(rl):
            s += float(sd['ratings'].iloc[i][key])
        ratings[key] = s / rl

    # word cloud
    wordcloud = tagWordCloud(words)
    ax1[index].imshow(wordcloud)
    ax1[index].axis('off')
    ax1[index].text(0.5, -0.15, speaker, size=15, ha="center", transform=ax1[index].transAxes)
    ax1[index].text(0.5, -0.45, '\n'.join(sd['title'].tolist()), size=15, ha="center", transform=ax1[index].transAxes)

    # bar chart
    labels = np.arange(len(keys))
    sizes = ratings.values()
    ax[index].barh(labels, sizes, color=colors, align='center')
    ax[index].set_yticks(labels)
    ax[index].set_yticklabels(keys)
```
